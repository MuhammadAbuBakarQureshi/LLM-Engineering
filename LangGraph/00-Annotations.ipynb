```python
from typing import TypedDict, Union, Optional, Any
```

## Union

```python
def square(x: Union[int, float]) -> Union[int, float]:

    return x * x

square(2)
```

## Optional

```python
def welcome_message(name: Optional[str] = None):

    if name is None:

        print("Welcome, random person")

    else: 

        print(f"Welcome, {name}")

welcome_message()
```

## Any

```python
def print_value(x: Any):

    print(x)

print_value("My name is bakar qureshi")
```

## Lambda

```python
square = lambda x: x * x
square(10)
```

```python
nums = [1, 2, 3, 4]
list(map(lambda num: num + 1, nums))
```
