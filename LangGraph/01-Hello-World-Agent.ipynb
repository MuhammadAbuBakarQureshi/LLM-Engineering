```python
from typing import TypedDict
from langgraph.graph import StateGraph
from IPython.display import Image, display
```

## Create Agent state (state_schema) and node as well

```python
class AgentState(TypedDict):
    message: str



def greeting_node(state: AgentState) -> AgentState:

    """Simple node that adds a greeting message to the state"""

    state['message'] = f"Hey, {state['message']}, how are you"

    return state
```

## Build graph

```python
graph = StateGraph(AgentState)

graph.add_node("greeter", greeting_node)

graph.set_entry_point("greeter")
graph.set_finish_point("greeter")

app = graph.compile()
```

## Draw graph

```python
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"message": "bakar"})
```

```python
result['message']
```

# Exercise -- Personalized Compliment Agent

```python
class AgentState(TypedDict):
    name: str



def give_compliment(state: AgentState) -> AgentState:
    """Gives compliment"""

    state['name'] = f"{state['name']}, you are doing an amazing job learning LangGrpah!"

    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("complimenter", give_compliment)
graph.set_entry_point("complimenter")
graph.set_finish_point("complimenter")

app = graph.compile()
```

```python
display(Image(app.get_graph().draw_mermaid_png()))

display(Image(app.get_graph().draw_png()))
```

```python
result = app.invoke({"name": "Bakar Qureshi"})
```
